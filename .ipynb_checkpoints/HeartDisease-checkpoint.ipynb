```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats librayr from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
%config IPCompleter.greedy=True
```

```python
# ageage in years
# sex(1 = male; 0 = female)
# cpchest pain type cp: chest pain type 
# -- Value 1: typical angina 
# -- Value 2: atypical angina 
# -- Value 3: non-anginal pain 
# -- Value 4: asymptomatic 
# trestbpsresting blood pressure (in mm Hg on admission to the hospital)
# cholserum cholestoral in mg/dl
# fbs(fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
# restecgresting electrocardiographic results
# thalachmaximum heart rate achieved
# exangexercise induced angina (1 = yes; 0 = no)
# oldpeakST depression induced by exercise relative to rest
# slopethe slope of the peak exercise ST segment
# canumber of major vessels (0-3) colored by flourosopy
# thal3 = normal; 6 = fixed defect; 7 = reversable defect
# target 0-4
```

```python
df_clv=pd.read_csv('cleveland.csv',header = None)
columns_name=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
df_clv.columns=columns_name
df_clv['city']='cleveland'
```

```python

df_clv.head()
```

```python
df_hun=pd.read_csv('hungarian.csv',header = None)

df_hun.columns=columns_name
df_hun['city']='hungarian'
df_hun.replace('?',-9,inplace=True)
```

```python
df_hun.head()
```

```python
df_va=pd.read_csv('va.csv',header = None)
df_va.columns=columns_name
df_va['city']='va'
df_va.replace('?',-9,inplace=True)
```

```python
df_va.head()
```

```python
df_sw=pd.read_csv('switzerland.csv',header = None)

df_sw.columns=columns_name
df_sw['city']='switzerland'
df_sw.replace('?',-9,inplace=True)
```

```python
df_sw.head()
```

```python
df=df_clv.append([df_hun,df_va,df_sw])
```
